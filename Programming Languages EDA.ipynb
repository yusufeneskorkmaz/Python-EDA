# Cleaning

```python
df[df.duplicated()]
```

```python
df.isnull().sum()
```

```python
numeric_columns = df.select_dtypes(include=['int64', 'float64'])
object_columns = df.select_dtypes(include=['object'])
```

```python
df[numeric_columns.columns] = df[numeric_columns.columns].fillna(0)
df[object_columns.columns] = df[object_columns.columns].fillna("No")
```

```python
df.drop(df.columns[12:43], axis=1, inplace=True)

df
```

```python
df.drop(['website', 'isbndb', 'reference', 'central_package_repository_count'], axis=1, inplace=True)

df
```

# Analysis

```python
df.sort_values(by='number_of_users', ascending=False)
```

```python
top_10_users = df[['title', 'number_of_users']].groupby('title').sum()
top_10_users = top_10_users.nlargest(10, 'number_of_users')

top_10_users
```

```python
top_10_jobs = df[['title', 'number_of_jobs']].groupby('title').sum()
top_10_jobs = top_10_jobs.nlargest(10, 'number_of_jobs')

top_10_jobs
```

```python
number_of_books = df[['title', 'book_count']].groupby('title').sum()
number_of_books = number_of_books.nlargest(10, 'book_count')

number_of_books
```

# Visualization

```python
colors = plt.cm.autumn(np.linspace(0, 1, len(top_10_users)))
```

```python
plt.figure(figsize=(10, 6))
plt.bar(top_10_users.index, top_10_users['number_of_users'], color=colors)
plt.xlabel('Programming Language')
plt.ylabel('Number of Users')
plt.title('Top 10 Software by Number of Users')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
plt.show()
```

```python
colors2 = plt.cm.winter(np.linspace(0, 1, len(top_10_jobs)))
```

```python

plt.figure(figsize=(10, 6))
plt.bar(top_10_jobs.index, top_10_jobs['number_of_jobs'], color=colors2)
plt.xlabel('Programming Language')
plt.ylabel('Number of jobs')
plt.title('Top 10 Software by Number of Jobs')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
plt.show()
```

```python
colors3 = plt.cm.summer(np.linspace(0, 1, len(number_of_books)))
```

```python
plt.figure(figsize=(10, 6))
plt.bar(number_of_books.index, number_of_books['book_count'], color=colors3)
plt.xlabel('Programming Language')
plt.ylabel('Number of Books')
plt.title('Number of  Books per Software')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
plt.show()
```
